```python
import numpy as np
import matplotlib.pyplot as plt

try:
    logo = np.random.rand(50, 50)

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("DOT")
    plt.grid(False)  # use False instead of 'off'
    
    
    
    dark_logo=1-logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("dark mode")
    plt.grid()
    plt.show()

except FileNotFoundError:
    print(" logo file not found")
```
